```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15            
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
### 남자 가지고 와서 찍기
df = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, nrows=1, usecols="B,E:Y", index_col='행정기관')

### 여자 가지고 와서 찍기
df_w = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, nrows=1, usecols="B,AB:AV", index_col='행정기관')
```

```python
df.index = df.index.str.replace(" ","",)
df_w.index = df_w.index.str.replace(" ","",)
```

```python
# 남자 나이 정리
df.index = df.index.str.replace(" ","",)
df.loc['전국'] = df.loc['전국'].str.replace(",","").astype(int)

# 여자 나이 정리
df_w.index = df_w.index.str.replace(" ","",)
df_w.loc['전국'] = df_w.loc['전국'].str.replace(",","").astype(int)
```

```python
# 남자컬럼 여자컬럼을 복사
df.columns = df_w.columns
```

```python
# 막대그래프로 출력하시오
plt.figure(figsize=(15,10))
# plt.xlim(1000,2300000)
plt.barh(df.columns,df.loc['전국']//10000, label='남자')
plt.barh(df_w.columns,-df_w.loc['전국']//10000, label='여자')
plt.title('2024년 인구 현황')
plt.ylabel('연령대')
plt.xlabel('인구수 (만 명)')

plt.legend()
plt.show()
```
