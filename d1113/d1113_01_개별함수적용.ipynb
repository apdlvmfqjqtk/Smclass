```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15            
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
plt.title('학생성적 그래프')
plt.plot(df.index, df['국어'], label='국어')
plt.plot(df.index, df['영어'], label='영어')
plt.plot(df.index, df['수학'], label='수학')

for i in range(8):
  plt.text(df.index[i], df['국어'][i]+0.6, df['국어'][i], ha='center')
  plt.text(df.index[i], df['영어'][i]+0.6, df['영어'][i], ha='center')
  plt.text(df.index[i], df['수학'][i]+0.6, df['수학'][i], ha='center')

plt.legend()
plt.show()
```

```python
def func(x):
  if x-30 < 0:
    return 0 
  return x-30
```

```python
# 누적막대그래프
plt.figure(figsize=(8,4))
plt.title('학생성적그래프')
plt.bar(df['이름'], df['국어'], width=0.6, label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.6, label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], width=0.6, label='수학')
n_list = df['국어'].apply(func)

for i in range(8):
  plt.text(df['이름'][i], n_list[i], df['국어'][i], ha='center')
  # plt.text(df['이름'][i], df['국어'][i]-30, df['국어'][i], ha='center')
  plt.text(df['이름'][i],df['국어'][i] + df['영어'][i]-30, df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + df['수학'][i]-30, df['수학'][i], ha='center')

# plt.legend(ncol=3)  # 가로로 배치
plt.legend()
plt.show()
```
