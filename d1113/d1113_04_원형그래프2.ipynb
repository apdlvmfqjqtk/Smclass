```python
## 원 그래프
values = [30, 25, 20, 13, 10, 2] # 100% 맞춘거
# values = [1, 1, 1, 1, 2, 1] # 비율 맞춰서 계산
explode = [0.1, 0, 0, 0, 0, 0]
labels = ['python','java','javascript','oracle','html','etc']
colors = ['plum', 'beige', 'skyblue', 'bisque', 'khaki', 'peachpuff']
# # webgeprops : 도넛 형태, edgecolor : 테두리 색상, linewidth : 테두리 두께
webgeprops = {'width': 0.65, 'edgecolor':'white', 'linewidth':2}

### 데이터 값이 2% 이하의 값은 출력되지 않도록 함수 적용
### pct : values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     # return ('{:.1f}%'.format(pct))
#     return (f'{pct:.1f}%')
#   else:
#     return ''

def custom_autopct(pct):
  return f'{pct:.1f}%' if pct >= 10 else f'{pct:.0f}'

# # wedgeprops : 중간 공백 , pctdistance : 글자 위치
plt.pie(values, labels=labels, colors=colors, startangle=90, counterclock=False, autopct=custom_autopct, \
         wedgeprops=webgeprops, pctdistance=0.7)

plt.show()
```
