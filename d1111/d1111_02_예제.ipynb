```python
import pandas as pd
import numpy as np

df = pd.read_csv('chipotle.tsv',sep='\t')
df.head(2)
```

```python
## 함수를 사용해서 
## item_price $ 부분을 없애고, float타입으로 변경해서 입력하시오.

def func(x):
 return float(x[1:])

df['item_price'] = df['item_price'].apply(func)
df['item_price'] = df['item_price'].apply(lambda x:float(x[1:]))
df.head(3)
```
