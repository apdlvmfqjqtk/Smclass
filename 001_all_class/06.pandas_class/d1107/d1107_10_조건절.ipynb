조건절

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 조건 (if문과 비슷)
df['키']>=185
```

```python
# 조건식을 변수에 저장하는 방식
filt = df['키'] >= 185 
df[filt]  # 조건에 해당되는 값만 출력
df[~filt] # 조건에 해당되지 않는 값만 출력 (~ : not)

# 직접 값을 넣는 방식
df[df['키'] >= 185]
df[~(df['키'] >= 185)]
```

```python
# 같은 결과값.
df[df['키']>185]          # 키가 185이상인 모든 칼럼 출력
df.loc[df['키'] >= 185]   # 키가 185이상인 모든 칼럼 출력, 칼럼을 지정할 수 있음
##-----
df.loc[df['키'] >= 185,'이름'] #키가 185 이상인 모든 '이름'컬럼만 출력
df.loc[df['키'] >= 185,['이름','키','SW특기']] 
df['키'] >= 185         # True, False의 값으로 출력
```

```python
# 조건신 and &, or
filt = (df['키'] >= 185) & (df['학교'] == '구로고')
df[filt]
```

```python
df.loc[(df['키']<170) | (df['학교'] == '구로고')]
```

문자열 조건 (STR)

```python
# 이름이 '강'으로 시작하는 사람 찾기 -> startwith("강")
# 문자열 함수 endwith, strip, replace
filt = df['이름'].str.startswith('강')
df[filt]
df[df['이름'].str.startswith('김')]

### '수' 가 포함된 이름을 검색 : contains()
filt = df['이름'].str.contains('수')
df[filt]

# lower: 소문자 변환
df['SW특기'].str.lower()

# upper: 대문자 변환
df['SW특기'].str.upper()

df['SW특기']
langs = ['Python','Java']
# isin리스트에 있는 단어가 포함돼 있는지 체크
filt = df['SW특기'].isin(langs) # 포함하는지 체크
df[filt]

# 소문자 취환
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

```

```python
# # Java라는 글자가 있는지 체크하려면
##  na = True: NaN True 인식, na= False: NaN False인식
df['SW특기'].str.contains('Java')
filt = df['SW특기'].str.contains('Java',na=False)
df[filt]

df.info()
```
