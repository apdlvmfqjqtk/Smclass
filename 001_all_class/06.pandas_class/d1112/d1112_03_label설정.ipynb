```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac 용
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## x,y축의 데이터
x = [1, 2, 3]  # 리스트, 튜플, DataFrame데이터. - 리스트 형태면 무조건 됨
y = [2, 4, 8]

## 선그래프 생성
plt.plot(x,y)

# 상단 타이틀
 # size: 글자 크기 변경
plt.title("라인그래프 Line", size=30)

# 그래프 X축 설명 colo: 글자색 변경/색상값 16코드 loc + left, center, right 글자위치
plt.xlabel('X축 날짜', color='#52bf6f', loc='left')

  # 그래프 Y축 설명 colo: 글자색 변경/색상값 16코드 loc + top, center, bottom 글자위치
plt.ylabel('Y축 인원', color='#152045', loc='top')

## 그래프 눈금단위
plt.xticks([1, 2, 3])
plt.yticks([2, 4, 8, 10])

plt.show()
```
