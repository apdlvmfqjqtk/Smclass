```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac 용
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## x,y축의 데이터
x = [1, 2, 3]  # 리스트, 튜플, DataFrame데이터. - 리스트 형태면 무조건 됨
y = [2, 4, 8]
yy = [1, 8, 3]
## 선그래프 생성
# linewidth: 선 두께, color: 선색상
plt.plot(x,y, label='인원변동', linewidth=5, marker='o', markersize=10, markeredgecolor='red', linestyle='-.')
# marker : 지점 표시- o, x, v, 3  markersize: 마커 크기, markeredgecolor : 마커 테두리색상, markerfacecolor: 마커 색상
# alpha: 투명도 설정, ls: 라인 스타일 :점선 -- dash선 -. dash점선
# plt.plot(x,yy, label='비용변동', marker='v', ms=10, mec='blue', mfc='green', alpha=0.4, ls=":")

# b: 선 색상, v: 마커표시, :선 모양
## 설정별 단축 표시
plt.plot(x, yy, 'bv:', label='비용변동')
# 상단 타이틀
 # size: 글자 크기 변경
plt.title("라인그래프 Line", size=30)

## 범례 : 그래프 선 설명문구. loc: 범례 위치 upper lower / left center right
# plt.legend(loc = 'lower center')
plt.legend(loc =(0.3,0.1))     # 좌표값으로 위치 설정 가능
```
